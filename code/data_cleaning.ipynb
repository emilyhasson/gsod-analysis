In this notebook, we will clean the 2000 data and prepare it for analysis. The same process can be applied across all years.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
year = 2000
```

```python
datapath = '/Users/samerens/Documents/GitHub/gsod-analysis/data/' # set to the path where you have the data stored
```

```python
df = pd.read_csv(datapath + str(year) + '_dir/' + str(year) + 'weatherdata.csv')
```

```python
df
```

The first thing we notice is a column called `Unnamed: 0` which appears to be a duplicate of the index column. Let's go ahead and drop it.

```python
df = df.drop('Unnamed: 0', axis = 1)
```

```python
df
```

Much better. Scrolling over to the right, we can see a lot of `NaN` values in the `SNDP`, `STP`, and `WBAN` columns. Let's take a closer look at what's going on.

```python
df.isna().sum()
```

```python
df.isna().sum().sum() / df.size
```

Holy smokes! 12% of the data is NaNs. We will drop the columns that have NaNs and keep the ones that don't.

```python
df = df.drop(df.columns[df.isna().sum() > 0], axis = 1)
```

```python
df
```

Much better. We still have 14 columns, so that should be enough to do some interesting analysis. We can come back later and try to find a more elegant way to handle the NaNs, perhaps through some sort of inputation. However, this is good enough for now.

One other thing we notice is some zeros that look out of place in the `temp_ft` column. While zero degrees Fahrenheit is a valid temperature, the placement of the values suggests that they may be filling in for missing data. Let's take a closer look at what's going on.

```python
df['date'] = df.apply(lambda row: pd.Timestamp(year = int(row['year']), month = int(row['month']), day = int(row['day'])), axis = 1)
```

```python
station_num = df['station_num'].unique()[0]
```

```python
sub_df = df[df['station_num'] == station_num]
```

```python
plt.figure()
plt.plot(sub_df['date'], sub_df['temp_ft'])
plt.show()
```

Indeed, the temperature appears to drop to zero and then spike back up again. Let's go ahead and drop all the rows where `temp_ft` is zero. Once again, we can come back later to develop a better approach if need be.

```python
df = df[df['temp_ft'] != 0]
```

```python
df
```

Much better, and we still have over 100K rows. Nothing else stands out as being obviously invalid, so let's go ahead ans save the data to a csv file so that we can analyze it later.

```python
df.to_csv(datapath + str(year) + '_dir/' + str(year) + 'weatherdataclean.csv', index = False)
```

And we're done!
