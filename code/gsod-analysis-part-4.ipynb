```python
warnings.filterwarnings(action = 'ignore')

stations = []
scores = []

for station in df3['station_num'].unique():
    
    try:
        
        sample_df = df[df['station_num'] == station]

        train_df, test_df = train_test_split(sample_df)

        train_x = train_df.drop('tornado', axis = 1)
        train_y = train_df['tornado']

        test_x = test_df.drop('tornado', axis = 1)
        test_y = test_df['tornado']

        model = LogisticRegression()
        model.fit(train_x, train_y)
        
        predict = model.predict(test_x)

        score = accuracy_score(y_true = test_y, y_pred = predict)

        stations.append(station)
        scores.append(score)
        
    except: continue
    
output_df = pd.DataFrame({'station': stations, 'score': score})
output_df.to_csv('/Users/samerens/Downloads/tornado_scores.csv', index = False)
```
